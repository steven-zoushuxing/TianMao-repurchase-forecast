    Hello 同学们，欢迎来到课程的练习部分，在本课程中，我们将通过三节练习完成一整套数据分析&建模过程。
    
    预测和防止客户流失对每个企业来说都是一个巨大的潜在收入来源，客户流失( Customer Churn)是指订阅某产品的客户终止了与公司的关系。客户流失的全部成本包括损失的收入和用新客户替换这些客户所涉及的营销成本，减少客户流失是每一家互联网公司的关键业务目标。
    
    在本项目中，我们将会针对腾讯系C端产品的用户流失情况进行分析，用户每天都会在产品上使用提供的服务与内容。部分用户会使用会员订阅模式，每月支付固定的会员费用，以升级服务。在这其中，用户可以随时升级、降级或取消该会员订阅服务。
    
    因此，我们的工作是对客户数据进行深度挖掘，并构建相应的模型来预测客户的流失率。

# 解决思路

这里的要求是**预测用户退订的概率**，用户产生流失的行为是一个“是or否”的问题，所以这里实际上是一个**分类问题**。

此外，我们还需要得知用户的退订概率，其本质上是一个排序问题，所以我们可以选择使用roc_auc_score进行模型评估。


- 解决流程

- 数据概况&清洗

    - 完成数据清洗
    - 完成多源数据整合（用户数据、操作日志等）
- 探索性数据分析
    - 探究退订用户的特征
- 特征工程
    - 基于基础数据，构造更多衍生特征
    - 将特征处理为模型可读样式
- 建模&优化
    - 模型选择分类模型，如逻辑回归、决策树等
    - 使用roc_auc_score对模型进行评估
    - 使用GridSearchCV对模型进行参数调优
- 总结&业务解读

# 特征工程

基于如上探索性数据分析，我们可以主要针对【用户粘性】&【用户习惯】这两个方面对用户构造如下特征：

- 用户粘性：
    - 用户浏览content的总量、丰富度；
    - 用户浏览author的总量、丰富度；
    - 用户访问日期间隔的最大、最小、均值；
    - 用户浏览连续访问日期最大、最小、均值；    
    - 用户单日访问次数的最大、最小、均值；
    
  
- 用户习惯：
    - 用户在不同weekday浏览的比例；
    - 用户在不同hour浏览的比例；
    - 用户浏览不同content占总浏览量比例的均值、最大值、标准差；
    - 用户浏览不同author占总浏览量比例的均值、最大值、标准差；
    

- 用户体验：
    - 用户浏览不同page的占比；
    
此外，保留如下用户特征：
- 性别
- 使用的浏览器及平台
- 使用的节点
- 所在省份

并将如上分类变量处理为哑变量

```python
# 浏览间隔的最大、最小、平均天数
continuous_days = df[['userId','date']].drop_duplicates()
tmp_days = continuous_days.groupby('userId')['date'].transform(lambda x:x.sort_values() - x.sort_values().shift()).dt.days
continuous_days['continuous_days'] = tmp_days
f_conti_days = continuous_days.groupby('userId')['continuous_days'].agg({'conti_days_mean':'mean',
                                                         'conti_days_max':'max',
                                                         'conti_days_min':'min'})
## 如果为空，则说明只有单天的数据，将其填充为999
f_conti_days = f_conti_days.fillna(999)

# 最后一天与第一天的间隔天数
f_lag_days = pd.Series(continuous_days.groupby('userId')['date'].apply(lambda x:x.max()-x.min()).dt.days,name='lag_days')
                     
```

## 特征筛选

- 查看特征之间的相关系数

```python
num_cols = [col for col in data.columns if col not in cat_cols+['Churn']]
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(X[num_cols].corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax);
```

```python
drop_cols = []
for i in range(len(num_cols)):
    for j in range(i+1,len(num_cols)):
        col = [num_cols[i],num_cols[j]]
        cor = X[col].corr().iloc[0,1]
        if abs(cor) >= 0.9:
            print(col,' ',cor)
            drop_cols.append(col[1])
```

```python
X_train_1 = X_train.drop(drop_cols,axis=1)
X_test_1 = X_test.drop(drop_cols,axis=1)
```

结论：
- `lag_days`即用户第一次浏览与最后一次浏览的日期差，该变量对用户退订有最大的负向影响，即：用户订阅持续时间越长，则越有可能退订；
- `content_nunique`即用户浏览内容的丰富度，该变量对用户退订有次大的负向影响，即：用户浏览（关注）的content越多越杂，则越有可能退订；
- `author_nunique`即用户浏览作者的丰富度，该变量对用户退订有最大的正向影响，即：用户浏览（关注）的author越丰富，则越没可能退订；

我们可以发现，退订用户浏览的内容越杂，说明该用户并未找到适合自己或者中意的内容，则越有可能发生退订行为；反之，若浏览的作者丰富度越高，则越容易吸引用户。
