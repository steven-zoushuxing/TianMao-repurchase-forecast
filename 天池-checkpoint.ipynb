# --------------4.特征工程--------------

```python
tmp=user_log_data.sample(1000)
```

## --------------4.1商品特征--------------

商品热度

    某商品发生交互的用户量（各个行为的数量，和总量）
    
    某商品发生交互行为的天数
    
商品广泛度

    某商品在多少家店铺出现过

```python
#针对某一个商品其每个交互行为用户的数量
tmp_with_user_nunique=tmp.groupby(['item_id','action_type'])['user_id'].nunique().unstack()
tmp_with_user_nunique=tmp_with_user_nunique.rename(columns=lambda x: f'item_user_nuique_{x}')
#填充缺失值
tmp_with_user_nunique=tmp_with_user_nunique.fillna(0)
```

```python
tmp_with_user_nunique
```

```python
#计算对于不同商品的用户总量
tmp.groupby('item_id')['user_id'].nunique()
#把用户总量转化为series类型加入原数据表中
tmp_with_user_nunique=tmp_with_user_nunique.join(pd.Series(tmp.groupby('item_id')['user_id'].nunique(),name='sum'))
tmp_with_user_nunique
```

```python
#商品发生交互行为的天数
tmp_with_tms_nunique=pd.Series(tmp.groupby('item_id')['time_stamp'].nunique(),name='item_tms_nunique')
#商品在多少家店铺出现过
tmp_with_seller_nunique=pd.Series(tmp.groupby('item_id')['seller_id'].nunique(),name='item_seller_nunique')
```

```python
item_feat=tmp_with_user_nunique.join(tmp_with_tms_nunique).join(tmp_with_seller_nunique)
item_feat=item_feat.reset_index()
```

```python
def get_normal_feats(data,base_feat):
    tag=base_feat.split('_')[0]
    #交互的用户量
    tmp_with_user_nunique=data.groupby([base_feat,'action_type'])['user_id'].nunique().unstack()
    tmp_with_user_nunique=tmp_with_user_nunique.rename(columns=lambda x:f'{tag}_user_nunique_{x}')
    #交互的用户总量
    tmp_with_user_nunique=tmp_with_user_nunique.join(pd.Series(data.groupby(base_feat)['user_id'].nunique(),name=f'{tag}_user_nunique_sum'))
    #交互的天数
    tmp_with_tms_nunique=pd.Series(tmp.groupby(base_feat)['time_stamp'].nunique(),name=f'{tag}_tms_nunique')
    item_feat=tmp_with_user_nunique.join(tmp_with_tms_nunique)
    #交互的店铺数,对于tag为店铺时不需要计算
    if tag!='seller':
        tmp_with_seller_nunique=pd.Series(tmp.groupby(base_feat)['seller_id'].nunique(),name=f'{tag}_seller_nunique')
        item_feat=item_feat.join(tmp_with_seller_nunique)
    #如果tag是类目则需要计算不同类目下包含多少种商品
    if tag!='item':
        tmp_with_contains=pd.Series(tmp.groupby(base_feat)['item_id'].nunique(),name=f'{tag}_item_contains')
        item_feat=item_feat.join(tmp_with_contains)
    item_feat=item_feat.reset_index()
    item_feat=item_feat.fillna(0)
    return item_feat
```

广泛度

丰富度

## --------------4.4用户交互特征--------------

用户对某xx的喜爱程度

    用户对某商品，类目，品牌，店铺的交互次数（总次数和各个action的次数）
    
    用户对某商品，类目，品牌，店铺的交互的天数
    
    商品，类目，品牌

用户的活跃度

    用户产生的总的交互次数
    
    用户交互过的商品，店铺的数量
    
    活跃的天数

```python
#定义用户对某商品，类目，品牌，店铺的喜爱程度，计算用户对于某商品，类目，品牌，店铺发生不同交互行为的次数，和所有交互行为的总次数及天数
def get_user_love_feat(data,base_feat):
    tag=base_feat.split('_')[0]
    #交互次数
    unique_action_count=data.groupby(['user_id',base_feat])['action_type'].value_counts().unstack()
    unique_action_count=unique_action_count.rename(columns=lambda x:f'{tag}_action_{x}')
    unique_action_count=unique_action_count.fillna(0)
    #交互次数的和
    unique_action_count[f'{tag}_all_action']=unique_action_count.sum(axis=1)
    #交互的天数
    user_tag_tms=pd.Series(data.groupby(['user_id',base_feat])['time_stamp'].nunique(),name=f'user_{tag}_tms_nunique')
    user_love=unique_action_count.join(user_tag_tms).reset_index()
    return user_love
```

```python
user_item_love=get_user_love_feat(tmp,'item_id')
user_cat_love=get_user_love_feat(tmp,'cat_id')
user_brand_love=get_user_love_feat(tmp,'brand_id')
user_seller_love=get_user_love_feat(tmp,'seller_id')
```

```python
user_item_love
```

```python
#用户活跃度
user_active=tmp.groupby('user_id')['action_type'].value_counts().unstack()
user_active=user_active.rename(columns=lambda x:f'user_action_{x}')
user_active=user_active.fillna(0)
user_active['user_all_action']=user_active.sum(axis=1)
```

```python
user_active
```

```python
user_interact=tmp.groupby('user_id')[['seller_id','item_id','cat_id','brand_id','time_stamp']].nunique()
user_interact=user_interact.rename(columns=lambda x:f'user_inter_{x}')
```

```python
user_interact
```

```python
user_act=user_active.join(user_interact)
user_act=user_act.reset_index()
user_act
```

```python
#定义用户的活跃度，计算用户对不同交互行为的次数及用户访问店铺的个数，用户访问商品的个数，用户访问类别的个数，用户访问品牌的个数，用户访问时间的个数
def get_user_active(data):
    user_active=data.groupby('user_id')['action_type'].value_counts().unstack()
    user_active=user_active.rename(columns=lambda x:f'user_action_{x}')
    user_active=user_active.fillna(0)
    user_active['user_all_action']=user_active.sum(axis=1)
    user_interact=data.groupby('user_id')[['seller_id','item_id','cat_id','brand_id','time_stamp']].nunique()
    user_interact=user_interact.rename(columns=lambda x:f'user_inter_{x}')
    user_act=user_active.join(user_interact)
    user_act=user_act.reset_index()
    return user_act
```
